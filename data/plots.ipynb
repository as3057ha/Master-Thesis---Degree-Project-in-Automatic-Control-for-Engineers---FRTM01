```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

# Plot: Feeding Repeatability

For one, two and three revolutions. Test the following:

- Different weigths and different speeds.

Measure consistancy, see if you can fit the results to a curve. 


```python


# Load data from the first sheet of the Excel file, skipping the first two rows and setting the third row as the header
df = pd.read_excel('Feeding Accuracy 0415.xlsx', sheet_name=0, skiprows=2)

# Remove unnamed columns
df = df.dropna(axis=1, how='all')

# Plotting Error vs Pulse Delay
plt.figure(figsize=(10, 6))
plt.scatter(df['Pulse Delay [ms]'], df['Error [mm]'])
plt.xlabel('Pulse Delay [ms]')
plt.ylabel('Error [mm]')
plt.title('Error vs Pulse Delay (One Revolution)')

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(df['Pulse Delay [ms]'], df['Error [mm]'])

# Plot the regression line
x = np.array([min(df['Pulse Delay [ms]']), max(df['Pulse Delay [ms]'])])
plt.plot(x, slope * x + intercept, color='red', label=f'Linear Regression (R={r_value:.2f})')

plt.legend()
plt.grid(True)
plt.show()

# Plotting Error vs Weight
plt.figure(figsize=(10, 6))
plt.scatter(df['Weight [g]'], df['Error [mm]'])
plt.xlabel('Weight [g]')
plt.ylabel('Error [mm]')
plt.title('Error vs Weight (One Revolution)')

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(df['Weight [g]'], df['Error [mm]'])

# Plot the regression line
x = np.array([min(df['Weight [g]']), max(df['Weight [g]'])])
plt.plot(x, slope * x + intercept, color='red', label=f'Linear Regression (R={r_value:.2f})')

plt.legend()
plt.grid(True)
plt.show()

```
